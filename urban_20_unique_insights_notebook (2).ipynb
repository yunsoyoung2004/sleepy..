
# 20 Unique Urban Insights Notebook

This notebook is structured as:

1. Insight Question  
2. Why It Matters  
3. Analysis Code  
4. Visualization Code  

Each section produces:
- a unique insight
- a matching visualization
- HTML and/or PNG outputs

Dataset paths are fixed to your existing folders.


```python

# ============================================================
# SETUP
# ============================================================

import os
import pandas as pd
import numpy as np

import plotly.express as px
import plotly.graph_objects as go

import matplotlib.pyplot as plt

BASE_DIR = r"C:\Users\82108\Desktop\새 폴더 (8)"

INPUT_PATH = os.path.join(
    BASE_DIR,
    "diverse_mobility_subway_analysis_filtered",
    "08_compare_subway_total_filtered.csv"
)

MAPPING_PATH = os.path.join(
    BASE_DIR,
    "station_mapping_final",
    "cell_to_nearest_station_VALID_5km_MASTER_ID_FILTERED.csv"
)

SAVE_DIR = os.path.join(
    BASE_DIR,
    "INSIGHT_20_VISUALS"
)

os.makedirs(SAVE_DIR, exist_ok=True)

df = pd.read_csv(INPUT_PATH)
mapping = pd.read_csv(MAPPING_PATH)

coord = mapping[
    [
        "MASTER_STATION_ID",
        "station_lat",
        "station_lon"
    ]
].drop_duplicates()

df = df.merge(
    coord,
    on="MASTER_STATION_ID",
    how="left"
)

print(df.shape)
```


# Insight 1. Hidden Demand Centers

### Core Question
What urban phenomenon does `hidden_demand_score` reveal?

### Why This Matters
Top hidden mobility hubs

### Expected Interpretation
This visualization helps interpret latent urban structure from mobility + subway sensing together.


```python

# ============================================================
# Insight 1: Hidden Demand Centers
# ============================================================

target = df.sort_values(
    "hidden_demand_score",
    ascending=False
).head(50)

print(target[
    [
        "nearest_station",
        "nearest_line",
        "hidden_demand_score"
    ]
].head(20))

fig = px.scatter_mapbox(
    target,
    lat="station_lat",
    lon="station_lon",
    size="hidden_demand_score",
    color="hidden_demand_score",
    hover_name="nearest_station",
    hover_data=[
        "nearest_line"
    ],
    zoom=10,
    height=800,
    mapbox_style="carto-darkmatter",
    color_continuous_scale="Turbo"
)

fig.update_layout(
    title="Insight 1: Hidden Demand Centers"
)

html_path = os.path.join(
    SAVE_DIR,
    "insight_01_hidden_demand_score.html"
)

png_path = os.path.join(
    SAVE_DIR,
    "insight_01_hidden_demand_score.png"
)

fig.write_html(html_path)

try:
    fig.write_image(
        png_path,
        scale=3
    )
except Exception as e:
    print("PNG save failed:", e)

fig.show()

print("[SAVED]", html_path)
```
