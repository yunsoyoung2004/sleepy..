```python
# ============================================================
# BEAUTIFUL VISUALIZATION PACKAGE
# Mobility vs Subway Hidden Demand Analysis
# ============================================================

import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# 지도용
try:
    import folium
    from folium.plugins import MarkerCluster, HeatMap
    FOLIUM_AVAILABLE = True
except:
    FOLIUM_AVAILABLE = False

# ============================================================
# 0. PATH
# ============================================================

INPUT_DIR = r"C:\Users\82108\Desktop\새 폴더 (8)\all_period_compare_output"
OUTPUT_DIR = r"C:\Users\82108\Desktop\새 폴더 (8)\visualization_output"
os.makedirs(OUTPUT_DIR, exist_ok=True)

compare_total_path = os.path.join(INPUT_DIR, "07_compare_subway_total.csv")
compare_30min_path = os.path.join(INPUT_DIR, "07_compare_subway_30min.csv")
compare_daily_path = os.path.join(INPUT_DIR, "07_compare_subway_daily.csv")
robust_path = os.path.join(INPUT_DIR, "08_robust_hidden_demand_stations.csv")
cell_level_path = os.path.join(INPUT_DIR, "02_cell_level_with_station_all_period.csv")
station_mobility_path = os.path.join(INPUT_DIR, "03_station_mobility_all_period.csv")

# ============================================================
# 1. LOAD
# ============================================================

def read_csv_safe(path):
    for enc in ["utf-8-sig", "cp949", "euc-kr", "utf-8"]:
        try:
            return pd.read_csv(path, encoding=enc)
        except:
            continue
    raise ValueError(f"cannot read: {path}")

compare = read_csv_safe(compare_total_path)
compare30 = read_csv_safe(compare_30min_path)
comparedaily = read_csv_safe(compare_daily_path)
robust = read_csv_safe(robust_path)

try:
    cell_level = read_csv_safe(cell_level_path)
except:
    cell_level = None

print("[COMPARE TOTAL]", compare.shape)
print(compare.columns.tolist())

print("[ROBUST]", robust.shape)
print(robust.columns.tolist())

# ============================================================
# 2. COLUMN NORMALIZATION
# ============================================================

# 좌표 컬럼 확인
# compare 파일에 station_lat/lon이 없을 수도 있으므로,
# 있으면 지도 사용, 없으면 지도는 station_mobility/cell_level에서 보강 필요

for df in [compare, compare30, comparedaily, robust]:
    for c in [
        "surrounding_total_flow",
        "subway_total_activity",
        "mobility_norm",
        "subway_norm",
        "mobility_minus_subway_gap",
        "hidden_demand_score",
        "subway_overrepresentation_score",
        "mean_distance_m"
    ]:
        if c in df.columns:
            df[c] = pd.to_numeric(df[c], errors="coerce").fillna(0)

# station name fallback
if "nearest_station" not in compare.columns and "station_name" in compare.columns:
    compare["nearest_station"] = compare["station_name"]

if "nearest_line" not in compare.columns and "line" in compare.columns:
    compare["nearest_line"] = compare["line"]

# ============================================================
# 3. STYLE HELPER
# ============================================================

def savefig(name):
    path = os.path.join(OUTPUT_DIR, name)
    plt.tight_layout()
    plt.savefig(path, dpi=250, bbox_inches="tight")
    plt.show()
    print("saved:", path)

# ============================================================
# 4. SCATTER: SUBWAY VS MOBILITY
# ============================================================

plt.figure(figsize=(10, 8))

types = compare["station_type"].fillna("normal") if "station_type" in compare.columns else pd.Series(["normal"] * len(compare))

for t in types.unique():
    sub = compare[types == t]
    plt.scatter(
        sub["subway_total_activity"],
        sub["surrounding_total_flow"],
        s=45,
        alpha=0.65,
        label=t
    )

# diagonal reference
xmax = compare["subway_total_activity"].max()
ymax = compare["surrounding_total_flow"].max()
limit = max(xmax, ymax)
plt.plot([0, limit], [0, limit], linestyle="--", linewidth=1)

plt.xlabel("Subway total activity")
plt.ylabel("Surrounding mobility flow")
plt.title("Mobility Demand vs. Subway Activity")
plt.legend()
savefig("01_scatter_mobility_vs_subway.png")

# ============================================================
# 5. TOP HIDDEN DEMAND BAR
# ============================================================

top_hidden = compare.sort_values("hidden_demand_score", ascending=False).head(25)

plt.figure(figsize=(11, 9))
plt.barh(
    top_hidden["nearest_station"].astype(str)[::-1],
    top_hidden["hidden_demand_score"][::-1]
)
plt.xlabel("Hidden demand score")
plt.ylabel("Station")
plt.title("Top Hidden Demand Stations")
savefig("02_top_hidden_demand_stations.png")

# ============================================================
# 6. TOP SUBWAY OVERREPRESENTED BAR
# ============================================================

if "subway_overrepresentation_score" in compare.columns:
    top_over = compare.sort_values("subway_overrepresentation_score", ascending=False).head(25)

    plt.figure(figsize=(11, 9))
    plt.barh(
        top_over["nearest_station"].astype(str)[::-1],
        top_over["subway_overrepresentation_score"][::-1]
    )
    plt.xlabel("Subway overrepresentation score")
    plt.ylabel("Station")
    plt.title("Top Subway-Overrepresented Stations")
    savefig("03_top_subway_overrepresented_stations.png")

# ============================================================
# 7. GAP DISTRIBUTION
# ============================================================

plt.figure(figsize=(10, 6))
plt.hist(compare["mobility_minus_subway_gap"], bins=40, alpha=0.8)
plt.axvline(0, linestyle="--", linewidth=1)
plt.xlabel("Mobility - Subway normalized gap")
plt.ylabel("Number of stations")
plt.title("Distribution of Mobility–Subway Discrepancy")
savefig("04_gap_distribution.png")

# ============================================================
# 8. STATION TYPE COUNTS
# ============================================================

if "station_type" in compare.columns:
    type_counts = compare["station_type"].value_counts()

    plt.figure(figsize=(8, 6))
    plt.bar(type_counts.index.astype(str), type_counts.values)
    plt.xticks(rotation=30, ha="right")
    plt.ylabel("Number of stations")
    plt.title("Station Type Classification")
    savefig("05_station_type_counts.png")

# ============================================================
# 9. DISTANCE VS HIDDEN DEMAND
# ============================================================

if "mean_distance_m" in compare.columns:
    plt.figure(figsize=(10, 7))
    plt.scatter(
        compare["mean_distance_m"],
        compare["hidden_demand_score"],
        s=55,
        alpha=0.65
    )
    plt.xlabel("Mean distance to station-linked cells (m)")
    plt.ylabel("Hidden demand score")
    plt.title("Distance-Aware Hidden Demand Pattern")
    savefig("06_distance_vs_hidden_demand.png")

# ============================================================
# 10. ROBUST HIDDEN DEMAND BAR
# ============================================================

if len(robust) > 0:
    robust_top = robust.sort_values("hidden_demand_score", ascending=False).head(30)

    name_col = "nearest_station" if "nearest_station" in robust_top.columns else "station_name"

    plt.figure(figsize=(11, 10))
    plt.barh(
        robust_top[name_col].astype(str)[::-1],
        robust_top["hidden_demand_score"][::-1]
    )
    plt.xlabel("Robust hidden demand score")
    plt.ylabel("Station")
    plt.title("Robust Hidden Demand Stations Across Subway Data Sources")
    savefig("07_robust_hidden_demand_stations.png")

# ============================================================
# 11. CORRELATION HEATMAP
# ============================================================

corr_cols = [
    "surrounding_total_flow",
    "subway_total_activity",
    "mobility_norm",
    "subway_norm",
    "mobility_minus_subway_gap",
    "hidden_demand_score",
    "subway_overrepresentation_score",
    "mean_distance_m",
    "mapped_cell_count",
    "active_cell_count"
]

corr_cols = [c for c in corr_cols if c in compare.columns]
corr = compare[corr_cols].corr()

plt.figure(figsize=(10, 8))
plt.imshow(corr, aspect="auto")
plt.colorbar(label="Correlation")
plt.xticks(range(len(corr_cols)), corr_cols, rotation=45, ha="right")
plt.yticks(range(len(corr_cols)), corr_cols)
plt.title("Correlation Matrix of Mobility–Subway Indicators")
savefig("08_correlation_heatmap.png")

# ============================================================
# 12. COMPARE 30MIN / DAILY / TOTAL ROBUSTNESS
# ============================================================

def prepare_rank(df, label):
    temp = df[["MASTER_STATION_ID", "nearest_station", "hidden_demand_score"]].copy()
    temp = temp.rename(columns={"hidden_demand_score": f"hidden_{label}"})
    return temp

rob = prepare_rank(compare30, "30min").merge(
    prepare_rank(comparedaily, "daily"),
    on=["MASTER_STATION_ID", "nearest_station"],
    how="inner"
).merge(
    prepare_rank(compare, "total"),
    on=["MASTER_STATION_ID", "nearest_station"],
    how="inner"
)

rob["hidden_mean"] = rob[["hidden_30min", "hidden_daily", "hidden_total"]].mean(axis=1)
rob["hidden_std"] = rob[["hidden_30min", "hidden_daily", "hidden_total"]].std(axis=1)

rob_top = rob.sort_values("hidden_mean", ascending=False).head(25)

plt.figure(figsize=(11, 8))
plt.errorbar(
    rob_top["hidden_mean"],
    rob_top["nearest_station"].astype(str),
    xerr=rob_top["hidden_std"],
    fmt="o",
    capsize=3
)
plt.xlabel("Mean hidden demand score across 30min / daily / total")
plt.ylabel("Station")
plt.title("Robustness of Hidden Demand Scores")
savefig("09_hidden_demand_robustness.png")

rob.to_csv(os.path.join(OUTPUT_DIR, "09_hidden_demand_robustness_table.csv"), index=False, encoding="utf-8-sig")

# ============================================================
# 13. CELL-LEVEL HEATMAP-LIKE SCATTER
# ============================================================

if cell_level is not None and {"CELL_X", "CELL_Y", "actual_total_flow"}.issubset(cell_level.columns):
    sample = cell_level.copy()
    sample["actual_total_flow"] = pd.to_numeric(sample["actual_total_flow"], errors="coerce").fillna(0)

    # 너무 많으면 상위 active cell만
    sample = sample.sort_values("actual_total_flow", ascending=False).head(30000)

    plt.figure(figsize=(10, 10))
    plt.scatter(
        sample["CELL_X"],
        sample["CELL_Y"],
        s=np.clip(sample["actual_total_flow"] / sample["actual_total_flow"].max() * 30, 1, 30),
        alpha=0.45
    )
    plt.xlabel("CELL_X (EPSG:5179)")
    plt.ylabel("CELL_Y (EPSG:5179)")
    plt.title("Cell-Level Mobility Intensity")
    savefig("10_cell_mobility_intensity.png")

# ============================================================
# 14. INTERACTIVE FOLIUM MAP
# ============================================================

# 지도는 station_lat/station_lon이 compare에 있어야 함
# 없으면 cell_level/station_mobility에서 보강 필요

lat_col = None
lon_col = None

for cand_lat, cand_lon in [
    ("station_lat", "station_lon"),
    ("lat", "lon"),
    ("latitude", "longitude")
]:
    if cand_lat in compare.columns and cand_lon in compare.columns:
        lat_col, lon_col = cand_lat, cand_lon
        break

if FOLIUM_AVAILABLE and lat_col is not None:
    m = folium.Map(
        location=[37.55, 126.98],
        zoom_start=10,
        tiles="CartoDB positron"
    )

    # hidden demand heat
    heat_data = []
    for _, row in compare.iterrows():
        score = row.get("hidden_demand_score", 0)
        lat = row[lat_col]
        lon = row[lon_col]
        if pd.notna(lat) and pd.notna(lon) and score > 0:
            heat_data.append([lat, lon, float(score)])

    if len(heat_data) > 0:
        HeatMap(
            heat_data,
            radius=22,
            blur=18,
            min_opacity=0.25
        ).add_to(m)

    cluster = MarkerCluster().add_to(m)

    top_map = compare.sort_values("hidden_demand_score", ascending=False).head(100)

    for _, row in top_map.iterrows():
        lat = row[lat_col]
        lon = row[lon_col]

        if pd.isna(lat) or pd.isna(lon):
            continue

        name = row.get("nearest_station", "")
        line = row.get("nearest_line", "")
        hidden = row.get("hidden_demand_score", 0)
        mobility = row.get("surrounding_total_flow", 0)
        subway = row.get("subway_total_activity", 0)
        stype = row.get("station_type", "")

        popup = f"""
        <b>{name}</b><br>
        Line: {line}<br>
        Type: {stype}<br>
        Hidden demand score: {hidden:.4f}<br>
        Mobility flow: {mobility:,.0f}<br>
        Subway activity: {subway:,.0f}
        """

        folium.CircleMarker(
            location=[lat, lon],
            radius=max(4, min(18, hidden * 20)),
            popup=folium.Popup(popup, max_width=300),
            fill=True,
            fill_opacity=0.75,
            opacity=0.9
        ).add_to(cluster)

    map_path = os.path.join(OUTPUT_DIR, "11_interactive_hidden_demand_map.html")
    m.save(map_path)
    print("saved:", map_path)
else:
    print("Folium map skipped: station_lat/station_lon not found or folium unavailable.")

# ============================================================
# 15. SUMMARY TABLE
# ============================================================

summary = {
    "n_compare_stations": len(compare),
    "n_hidden_stations": int((compare["station_type"] == "mobility_high_subway_low").sum()) if "station_type" in compare.columns else None,
    "n_subway_overrepresented": int((compare["station_type"] == "subway_high_mobility_low").sum()) if "station_type" in compare.columns else None,
    "top_hidden_station": compare.sort_values("hidden_demand_score", ascending=False).iloc[0]["nearest_station"],
    "top_hidden_score": compare.sort_values("hidden_demand_score", ascending=False).iloc[0]["hidden_demand_score"],
    "correlation_mobility_subway": compare["surrounding_total_flow"].corr(compare["subway_total_activity"])
}

summary_df = pd.DataFrame([summary])
summary_df.to_csv(os.path.join(OUTPUT_DIR, "00_visualization_summary.csv"), index=False, encoding="utf-8-sig")

print("\n================================================")
print("VISUALIZATION DONE")
print("================================================")
print("output folder:", OUTPUT_DIR)
print(summary_df)
```

```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm

# ============================================================
# 0. 경로
# ============================================================

mapping_path = r"C:\Users\82108\Desktop\새 폴더 (8)\station_mapping_final\cell_to_nearest_station_VALID_5km_MASTER_ID.csv"

analysis_path = r"C:\Users\82108\Desktop\새 폴더 (8)\masterid_subway_compare_20170101\04_masterid_mobility_vs_subway.csv"

output_dir = r"C:\Users\82108\Desktop\새 폴더 (8)\final_map_results"
os.makedirs(output_dir, exist_ok=True)

# ============================================================
# 1. 한글 폰트 설정
# ============================================================

font_candidates = [
    r"C:\Windows\Fonts\malgun.ttf",
    r"C:\Windows\Fonts\NanumGothic.ttf",
    r"C:\Windows\Fonts\NanumBarunGothic.ttf",
]

for font_path in font_candidates:
    if os.path.exists(font_path):
        font_name = fm.FontProperties(fname=font_path).get_name()
        plt.rcParams["font.family"] = font_name
        print("Korean font loaded:", font_name)
        break

plt.rcParams["axes.unicode_minus"] = False

# ============================================================
# 2. 데이터 로드
# ============================================================

mapping = pd.read_csv(mapping_path, encoding="utf-8-sig")
analysis = pd.read_csv(analysis_path, encoding="utf-8-sig")

mapping.columns = mapping.columns.str.strip()
analysis.columns = analysis.columns.str.strip()

print("[MAPPING]")
print(mapping.shape)
print(mapping.columns.tolist())

print("\n[ANALYSIS]")
print(analysis.shape)
print(analysis.columns.tolist())

# ============================================================
# 3. station 좌표 만들기
# mapping은 cell 단위라 station별 좌표만 중복 제거해서 사용
# ============================================================

station_geo = (
    mapping[[
        "MASTER_STATION_ID",
        "nearest_station",
        "nearest_line",
        "station_x_5179",
        "station_y_5179",
        "station_lat",
        "station_lon"
    ]]
    .drop_duplicates(subset=["MASTER_STATION_ID"])
    .copy()
)

station_geo["MASTER_STATION_ID"] = pd.to_numeric(
    station_geo["MASTER_STATION_ID"],
    errors="coerce"
).astype("Int64")

analysis["MASTER_STATION_ID"] = pd.to_numeric(
    analysis["MASTER_STATION_ID"],
    errors="coerce"
).astype("Int64")

# ============================================================
# 4. 분석 결과 + 좌표 merge
# ============================================================

df = analysis.merge(
    station_geo,
    on="MASTER_STATION_ID",
    how="left",
    suffixes=("", "_geo")
)

print("\n[MERGED]")
print(df.shape)
print("missing station_x:", df["station_x_5179"].isna().sum())
print("missing station_y:", df["station_y_5179"].isna().sum())

df = df.dropna(subset=["station_x_5179", "station_y_5179"]).copy()

# ============================================================
# 5. 수치형 정리
# ============================================================

num_cols = [
    "surrounding_total_flow",
    "subway_total_activity",
    "mobility_minus_subway_gap",
    "hidden_demand_score",
    "subway_overrepresentation_score",
    "mean_distance_m",
    "mapped_cell_count",
    "active_cell_count",
    "station_x_5179",
    "station_y_5179"
]

for c in num_cols:
    if c in df.columns:
        df[c] = pd.to_numeric(df[c], errors="coerce").fillna(0)

def scale_size(s, min_size=30, max_size=650):
    s = pd.to_numeric(s, errors="coerce").fillna(0)
    if s.max() == s.min():
        return pd.Series(min_size, index=s.index)
    return min_size + (s - s.min()) / (s.max() - s.min()) * (max_size - min_size)

# ============================================================
# 6. 지도 1: 생활이동 > 지하철 과소포착 지도
# ============================================================

hidden_df = df[df["station_type"] == "mobility_high_subway_low"].copy()

plt.figure(figsize=(10, 10))

plt.scatter(
    df["station_x_5179"],
    df["station_y_5179"],
    s=12,
    alpha=0.18,
    label="전체 비교 가능 역"
)

plt.scatter(
    hidden_df["station_x_5179"],
    hidden_df["station_y_5179"],
    s=scale_size(hidden_df["hidden_demand_score"], 100, 700),
    alpha=0.75,
    label="생활이동 높음 / 지하철 낮음"
)

for _, row in hidden_df.sort_values("hidden_demand_score", ascending=False).head(20).iterrows():
    plt.text(
        row["station_x_5179"],
        row["station_y_5179"],
        str(row["nearest_station"]),
        fontsize=8
    )

plt.title("생활이동 대비 지하철 과소포착 후보 역")
plt.xlabel("X 좌표 (EPSG:5179)")
plt.ylabel("Y 좌표 (EPSG:5179)")
plt.legend()
plt.axis("equal")
plt.tight_layout()

save_path = os.path.join(output_dir, "01_hidden_demand_station_map.png")
plt.savefig(save_path, dpi=300)
plt.show()

print("saved:", save_path)

# ============================================================
# 7. 지도 2: 지하철 > 생활이동 과대표현 지도
# ============================================================

over_df = df[df["station_type"] == "subway_high_mobility_low"].copy()

plt.figure(figsize=(10, 10))

plt.scatter(
    df["station_x_5179"],
    df["station_y_5179"],
    s=12,
    alpha=0.18,
    label="전체 비교 가능 역"
)

plt.scatter(
    over_df["station_x_5179"],
    over_df["station_y_5179"],
    s=scale_size(over_df["subway_overrepresentation_score"], 100, 700),
    alpha=0.75,
    label="지하철 높음 / 생활이동 낮음"
)

for _, row in over_df.sort_values("subway_overrepresentation_score", ascending=False).head(20).iterrows():
    plt.text(
        row["station_x_5179"],
        row["station_y_5179"],
        str(row["nearest_station"]),
        fontsize=8
    )

plt.title("지하철 activity 과대표현 후보 역")
plt.xlabel("X 좌표 (EPSG:5179)")
plt.ylabel("Y 좌표 (EPSG:5179)")
plt.legend()
plt.axis("equal")
plt.tight_layout()

save_path = os.path.join(output_dir, "02_subway_overrepresented_station_map.png")
plt.savefig(save_path, dpi=300)
plt.show()

print("saved:", save_path)

# ============================================================
# 8. 지도 3: 전체 gap 지도
# 양수 = 생활이동 우세 / 음수 = 지하철 우세
# ============================================================

plt.figure(figsize=(10, 10))

sc = plt.scatter(
    df["station_x_5179"],
    df["station_y_5179"],
    c=df["mobility_minus_subway_gap"],
    s=scale_size(np.log1p(df["surrounding_total_flow"]), 30, 500),
    cmap="coolwarm",
    alpha=0.78
)

plt.colorbar(sc, label="생활이동 - 지하철 정규화 격차")

plt.title("역별 생활이동-지하철 수요 격차 지도")
plt.xlabel("X 좌표 (EPSG:5179)")
plt.ylabel("Y 좌표 (EPSG:5179)")
plt.axis("equal")
plt.tight_layout()

save_path = os.path.join(output_dir, "03_mobility_subway_gap_map.png")
plt.savefig(save_path, dpi=300)
plt.show()

print("saved:", save_path)

# ============================================================
# 9. 지도 4: 역 반경/평균 연결거리 지도
# ============================================================

plt.figure(figsize=(10, 10))

sc = plt.scatter(
    df["station_x_5179"],
    df["station_y_5179"],
    c=df["mean_distance_m"],
    s=scale_size(df["hidden_demand_score"], 30, 550),
    cmap="viridis",
    alpha=0.78
)

plt.colorbar(sc, label="역에 연결된 cell 평균 거리(m)")

far_df = df.sort_values("mean_distance_m", ascending=False).head(20)

for _, row in far_df.iterrows():
    plt.text(
        row["station_x_5179"],
        row["station_y_5179"],
        str(row["nearest_station"]),
        fontsize=8
    )

plt.title("역 반경 기반 평균 연결거리 지도")
plt.xlabel("X 좌표 (EPSG:5179)")
plt.ylabel("Y 좌표 (EPSG:5179)")
plt.axis("equal")
plt.tight_layout()

save_path = os.path.join(output_dir, "04_station_mean_distance_map.png")
plt.savefig(save_path, dpi=300)
plt.show()

print("saved:", save_path)

# ============================================================
# 10. 해석용 테이블 저장
# ============================================================

summary_cols = [
    "MASTER_STATION_ID",
    "nearest_station",
    "nearest_line",
    "station_type",
    "mapped_cell_count",
    "active_cell_count",
    "mean_distance_m",
    "surrounding_total_flow",
    "subway_total_activity",
    "mobility_minus_subway_gap",
    "hidden_demand_score",
    "subway_overrepresentation_score",
    "station_x_5179",
    "station_y_5179",
    "station_lat",
    "station_lon"
]

summary = df[[c for c in summary_cols if c in df.columns]].copy()
summary = summary.sort_values("hidden_demand_score", ascending=False)

summary_path = os.path.join(output_dir, "00_map_interpretation_table.csv")
summary.to_csv(summary_path, index=False, encoding="utf-8-sig")

print("summary saved:", summary_path)
print("DONE")
```

```python
import os, glob, re, math
import pandas as pd
import numpy as np

# ============================================================
# 0. 경로
# ============================================================

mapping_path = r"C:\Users\82108\Desktop\새 폴더 (8)\station_mapping_final\cell_to_nearest_station_VALID_5km_MASTER_ID.csv"

mobility_dir = r"C:\Users\82108\Desktop\새 폴더 (7)\mobility_cell_output"

subway_30min_dir = r"C:\Users\82108\Desktop\새 폴더 (8)\30_od_output (1)\30_od_output"

subway_daily_dir = r"C:\Users\82108\Desktop\새 폴더 (8)\od_output\od_output"

output_dir = r"C:\Users\82108\Desktop\새 폴더 (8)\diverse_mobility_subway_analysis"
os.makedirs(output_dir, exist_ok=True)

# 선택: 숫자형 CELL_ID_BASE 복원용 cell master가 있으면 넣기
cell_master_path = r"C:\Users\82108\Desktop\새 폴더 (8)\CELL_250x250(UTMK_epsg 5179).csv"
USE_CELL_MASTER_FOR_NUMERIC_BASE = True

# ============================================================
# 1. 유틸
# ============================================================

def read_csv_safe(path):
    for enc in ["utf-8-sig", "cp949", "euc-kr", "utf-8"]:
        try:
            df = pd.read_csv(path, encoding=enc)
            df.columns = df.columns.astype(str).str.strip()
            print("read success:", os.path.basename(path), enc)
            return df
        except UnicodeDecodeError:
            pass
    raise ValueError("CSV 읽기 실패: " + path)

def normalize_minmax(s):
    s = pd.to_numeric(s, errors="coerce").fillna(0)
    if len(s) == 0 or s.max() == s.min():
        return pd.Series(0, index=s.index)
    return (s - s.min()) / (s.max() - s.min())

def safe_div(a, b):
    b = np.where(np.asarray(b) == 0, np.nan, b)
    out = np.asarray(a) / b
    return pd.Series(out).replace([np.inf, -np.inf], np.nan).fillna(0)

def is_korean_cell_id(series):
    return series.astype(str).str.contains(r"[가-힣]", regex=True)

def entropy_from_values(values):
    arr = np.array(values, dtype=float)
    arr = arr[arr > 0]
    if len(arr) == 0:
        return 0.0
    p = arr / arr.sum()
    return float(-(p * np.log(p)).sum())

def classify_period(dt):
    h = pd.to_datetime(dt, errors="coerce").dt.hour
    return np.select(
        [
            h.between(6, 9),
            h.between(10, 15),
            h.between(16, 20),
            h.between(21, 23) | h.between(0, 5),
        ],
        ["morning_peak", "daytime", "evening_peak", "night"],
        default="unknown"
    )

# ============================================================
# 2. mapping 로드
# ============================================================

mapping = read_csv_safe(mapping_path)

mapping["CELL_ID"] = mapping["CELL_ID"].astype(str).str.strip()
mapping["MASTER_STATION_ID"] = pd.to_numeric(mapping["MASTER_STATION_ID"], errors="coerce")
mapping = mapping.dropna(subset=["MASTER_STATION_ID"]).copy()
mapping["MASTER_STATION_ID"] = mapping["MASTER_STATION_ID"].astype(int)

needed_mapping = [
    "MASTER_STATION_ID", "CELL_ID", "CELL_X", "CELL_Y",
    "nearest_station", "nearest_line",
    "station_x_5179", "station_y_5179",
    "nearest_station_distance_m", "distance_group"
]
missing = [c for c in needed_mapping if c not in mapping.columns]
if missing:
    raise ValueError(f"mapping 필수 컬럼 없음: {missing}")

mapping = (
    mapping
    .sort_values(["CELL_ID", "nearest_station_distance_m", "MASTER_STATION_ID"])
    .drop_duplicates(subset=["CELL_ID"])
    .copy()
)

print("\n[MAPPING]")
print(mapping.shape)
print("unique cells:", mapping["CELL_ID"].nunique())
print("unique stations:", mapping["MASTER_STATION_ID"].nunique())

# ============================================================
# 3. 숫자형 CELL_ID_BASE 복원용 cell master
# ============================================================

cell_master = None
if USE_CELL_MASTER_FOR_NUMERIC_BASE and os.path.exists(cell_master_path):
    cell_master = read_csv_safe(cell_master_path)
    cell_master.columns = cell_master.columns.astype(str).str.strip()
    cell_master = cell_master.reset_index().rename(columns={"index": "CELL_INDEX"})
    cell_master["CELL_INDEX"] = cell_master["CELL_INDEX"].astype(int)
    cell_master["CELL_ID"] = cell_master["CELL_ID"].astype(str).str.strip()
    print("\n[CELL MASTER]")
    print(cell_master.shape)
else:
    print("\n[CELL MASTER] not used")

# ============================================================
# 4. mobility 전체 parquet 읽기 + CELL_ID 복원 + 시간대 다양성 집계
# ============================================================

mobility_files = sorted(glob.glob(os.path.join(mobility_dir, "**", "*.parquet"), recursive=True))
print("\n[MOBILITY FILES]", len(mobility_files))

mobility_parts = []
period_parts = []

for i, path in enumerate(mobility_files, 1):
    try:
        df = pd.read_parquet(
            path,
            columns=["datetime", "CELL_ID_BASE", "out_cnt", "in_cnt", "net_cnt"]
        )

        df["CELL_ID_BASE_RAW"] = df["CELL_ID_BASE"].astype(str).str.strip()
        korean_mask = is_korean_cell_id(df["CELL_ID_BASE_RAW"])

        # 한글 CELL_ID는 그대로 사용
        df_kor = df[korean_mask].copy()
        df_kor["CELL_ID"] = df_kor["CELL_ID_BASE_RAW"]

        # 숫자형 CELL_ID_BASE는 cell_master index로 복원
        if cell_master is not None:
            df_num = df[~korean_mask].copy()
            df_num["CELL_INDEX"] = pd.to_numeric(df_num["CELL_ID_BASE_RAW"], errors="coerce")
            df_num = df_num.dropna(subset=["CELL_INDEX"]).copy()
            df_num["CELL_INDEX"] = df_num["CELL_INDEX"].astype(int)

            df_num = df_num.merge(
                cell_master[["CELL_INDEX", "CELL_ID"]],
                on="CELL_INDEX",
                how="left"
            )
            df_num = df_num.dropna(subset=["CELL_ID"]).copy()

            df = pd.concat([df_kor, df_num], ignore_index=True)
        else:
            df = df_kor

        for c in ["out_cnt", "in_cnt", "net_cnt"]:
            df[c] = pd.to_numeric(df[c], errors="coerce").fillna(0)

        df["total_flow"] = df["out_cnt"] + df["in_cnt"]
        df["period"] = classify_period(df["datetime"])

        agg = (
            df.groupby("CELL_ID", as_index=False)
            .agg(
                mobility_out_sum=("out_cnt", "sum"),
                mobility_in_sum=("in_cnt", "sum"),
                mobility_net_sum=("net_cnt", "sum"),
                mobility_record_count=("datetime", "count")
            )
        )
        agg["source_file"] = os.path.basename(path)
        mobility_parts.append(agg)

        period_agg = (
            df.groupby(["CELL_ID", "period"], as_index=False)
            .agg(period_total_flow=("total_flow", "sum"))
        )
        period_agg["source_file"] = os.path.basename(path)
        period_parts.append(period_agg)

        if i % 100 == 0:
            print(f"loaded mobility {i}/{len(mobility_files)}")

    except Exception as e:
        print("skip mobility:", path)
        print("reason:", e)

if not mobility_parts:
    raise ValueError("읽힌 mobility parquet가 없음")

mobility_all = pd.concat(mobility_parts, ignore_index=True)

cell_mobility = (
    mobility_all
    .groupby("CELL_ID", as_index=False)
    .agg(
        mobility_out_sum=("mobility_out_sum", "sum"),
        mobility_in_sum=("mobility_in_sum", "sum"),
        mobility_net_sum=("mobility_net_sum", "sum"),
        mobility_record_count=("mobility_record_count", "sum"),
        mobility_file_count=("source_file", "nunique")
    )
)

cell_mobility["actual_total_flow"] = cell_mobility["mobility_out_sum"] + cell_mobility["mobility_in_sum"]
cell_mobility["actual_abs_net_flow"] = cell_mobility["mobility_net_sum"].abs()
cell_mobility["directional_imbalance"] = (
    cell_mobility["actual_abs_net_flow"] /
    (cell_mobility["actual_total_flow"] + 1e-9)
)

cell_mobility.to_csv(
    os.path.join(output_dir, "01_cell_mobility_all_period_restored.csv"),
    index=False,
    encoding="utf-8-sig"
)

# 시간대별 cell mobility
period_all = pd.concat(period_parts, ignore_index=True)
cell_period = (
    period_all.groupby(["CELL_ID", "period"], as_index=False)
    .agg(period_total_flow=("period_total_flow", "sum"))
)

cell_period_wide = (
    cell_period.pivot_table(
        index="CELL_ID",
        columns="period",
        values="period_total_flow",
        aggfunc="sum",
        fill_value=0
    )
    .reset_index()
)

for p in ["morning_peak", "daytime", "evening_peak", "night"]:
    if p not in cell_period_wide.columns:
        cell_period_wide[p] = 0

cell_period_wide["temporal_entropy"] = cell_period_wide[
    ["morning_peak", "daytime", "evening_peak", "night"]
].apply(lambda r: entropy_from_values(r.values), axis=1)

cell_period_wide.to_csv(
    os.path.join(output_dir, "01b_cell_temporal_profile.csv"),
    index=False,
    encoding="utf-8-sig"
)

print("\n[CELL MOBILITY]")
print(cell_mobility.shape)
print(cell_mobility.head())

# ============================================================
# 5. cell-level merge
# ============================================================

cell_level = mapping.merge(cell_mobility, on="CELL_ID", how="left")
cell_level = cell_level.merge(cell_period_wide, on="CELL_ID", how="left")

fill_cols = [
    "mobility_out_sum", "mobility_in_sum", "mobility_net_sum",
    "mobility_record_count", "mobility_file_count",
    "actual_total_flow", "actual_abs_net_flow",
    "directional_imbalance",
    "morning_peak", "daytime", "evening_peak", "night",
    "temporal_entropy"
]
for c in fill_cols:
    if c in cell_level.columns:
        cell_level[c] = pd.to_numeric(cell_level[c], errors="coerce").fillna(0)

cell_level.to_csv(
    os.path.join(output_dir, "02_cell_level_with_station_all_period_restored.csv"),
    index=False,
    encoding="utf-8-sig"
)

# ============================================================
# 6. station mobility 다양성 집계
# ============================================================

station_mobility = (
    cell_level
    .groupby(["MASTER_STATION_ID", "nearest_station", "nearest_line"], as_index=False)
    .agg(
        mapped_cell_count=("CELL_ID", "count"),
        active_cell_count=("actual_total_flow", lambda x: (x > 0).sum()),
        surrounding_total_flow=("actual_total_flow", "sum"),
        surrounding_outflow=("mobility_out_sum", "sum"),
        surrounding_inflow=("mobility_in_sum", "sum"),
        surrounding_netflow=("mobility_net_sum", "sum"),
        mean_cell_flow=("actual_total_flow", "mean"),
        median_cell_flow=("actual_total_flow", "median"),
        max_cell_flow=("actual_total_flow", "max"),
        mean_distance_m=("nearest_station_distance_m", "mean"),
        min_distance_m=("nearest_station_distance_m", "min"),
        max_distance_m=("nearest_station_distance_m", "max"),
        mean_directional_imbalance=("directional_imbalance", "mean"),
        morning_peak_flow=("morning_peak", "sum"),
        daytime_flow=("daytime", "sum"),
        evening_peak_flow=("evening_peak", "sum"),
        night_flow=("night", "sum"),
        mean_temporal_entropy=("temporal_entropy", "mean"),
    )
)

station_mobility["active_cell_ratio"] = (
    station_mobility["active_cell_count"] / station_mobility["mapped_cell_count"]
)

station_mobility["station_abs_net_flow"] = station_mobility["surrounding_netflow"].abs()
station_mobility["station_directional_imbalance"] = (
    station_mobility["station_abs_net_flow"] /
    (station_mobility["surrounding_total_flow"] + 1e-9)
)

# 시간대 dominance
period_cols = ["morning_peak_flow", "daytime_flow", "evening_peak_flow", "night_flow"]
station_mobility["dominant_period"] = station_mobility[period_cols].idxmax(axis=1)
station_mobility["dominant_period"] = station_mobility["dominant_period"].str.replace("_flow", "", regex=False)

station_mobility["period_entropy"] = station_mobility[period_cols].apply(
    lambda r: entropy_from_values(r.values), axis=1
)

# ============================================================
# 7. 거리권별 다양성 집계
# ============================================================

distance_band = (
    cell_level
    .groupby(["MASTER_STATION_ID", "distance_group"], as_index=False)
    .agg(
        band_cell_count=("CELL_ID", "count"),
        band_active_cell_count=("actual_total_flow", lambda x: (x > 0).sum()),
        band_total_flow=("actual_total_flow", "sum"),
        band_mean_flow=("actual_total_flow", "mean"),
        band_max_flow=("actual_total_flow", "max")
    )
)

distance_band.to_csv(
    os.path.join(output_dir, "03_distance_band_station_mobility.csv"),
    index=False,
    encoding="utf-8-sig"
)

# 거리권 entropy / outer share
band_wide = distance_band.pivot_table(
    index="MASTER_STATION_ID",
    columns="distance_group",
    values="band_total_flow",
    aggfunc="sum",
    fill_value=0
).reset_index()

for g in ["0-250m", "250-500m", "500m-1km", "1-2km", "2-5km"]:
    if g not in band_wide.columns:
        band_wide[g] = 0

band_cols = ["0-250m", "250-500m", "500m-1km", "1-2km", "2-5km"]
band_wide["distance_band_entropy"] = band_wide[band_cols].apply(
    lambda r: entropy_from_values(r.values), axis=1
)
band_wide["outer_1km_plus_flow"] = band_wide["1-2km"] + band_wide["2-5km"]
band_wide["total_band_flow"] = band_wide[band_cols].sum(axis=1)
band_wide["outer_1km_plus_share"] = (
    band_wide["outer_1km_plus_flow"] / (band_wide["total_band_flow"] + 1e-9)
)

station_mobility = station_mobility.merge(
    band_wide[["MASTER_STATION_ID", "distance_band_entropy", "outer_1km_plus_share"]],
    on="MASTER_STATION_ID",
    how="left"
)

station_mobility[["distance_band_entropy", "outer_1km_plus_share"]] = station_mobility[
    ["distance_band_entropy", "outer_1km_plus_share"]
].fillna(0)

station_mobility.to_csv(
    os.path.join(output_dir, "04_station_mobility_diverse_features.csv"),
    index=False,
    encoding="utf-8-sig"
)

print("\n[STATION MOBILITY DIVERSE]")
print(station_mobility.shape)
print(station_mobility.head())

# ============================================================
# 8. subway 전체 parquet 읽기 함수
# ============================================================

def load_subway_folder(folder, label):
    files = sorted(glob.glob(os.path.join(folder, "**", "*.parquet"), recursive=True))
    print(f"\n[{label} FILES]", len(files))

    parts = []
    period_parts = []

    for i, path in enumerate(files, 1):
        try:
            df = pd.read_parquet(path)

            if "subwayid" in df.columns and "STATION_ID" not in df.columns:
                df = df.rename(columns={"subwayid": "STATION_ID"})

            needed = ["datetime", "STATION_ID", "od_out_cnt", "od_in_cnt", "od_net_cnt"]
            missing = [c for c in needed if c not in df.columns]
            if missing:
                print("skip:", os.path.basename(path), "missing:", missing)
                continue

            df = df[needed].copy()
            df["STATION_ID"] = pd.to_numeric(df["STATION_ID"], errors="coerce")

            for c in ["od_out_cnt", "od_in_cnt", "od_net_cnt"]:
                df[c] = pd.to_numeric(df[c], errors="coerce").fillna(0)

            df = df.dropna(subset=["STATION_ID"]).copy()
            df["MASTER_STATION_ID"] = df["STATION_ID"].astype(int)
            df["subway_total_activity_row"] = df["od_out_cnt"] + df["od_in_cnt"]
            df["period"] = classify_period(df["datetime"])

            agg = (
                df.groupby("MASTER_STATION_ID", as_index=False)
                .agg(
                    subway_out_sum=("od_out_cnt", "sum"),
                    subway_in_sum=("od_in_cnt", "sum"),
                    subway_net_sum=("od_net_cnt", "sum"),
                    subway_record_count=("datetime", "count")
                )
            )
            agg["source_file"] = os.path.basename(path)
            parts.append(agg)

            pagg = (
                df.groupby(["MASTER_STATION_ID", "period"], as_index=False)
                .agg(subway_period_activity=("subway_total_activity_row", "sum"))
            )
            pagg["source_file"] = os.path.basename(path)
            period_parts.append(pagg)

            if i % 100 == 0:
                print(f"loaded {label} {i}/{len(files)}")

        except Exception as e:
            print("skip subway:", path)
            print("reason:", e)

    if not parts:
        return pd.DataFrame(), pd.DataFrame()

    all_df = pd.concat(parts, ignore_index=True)

    out = (
        all_df.groupby("MASTER_STATION_ID", as_index=False)
        .agg(
            subway_out_sum=("subway_out_sum", "sum"),
            subway_in_sum=("subway_in_sum", "sum"),
            subway_net_sum=("subway_net_sum", "sum"),
            subway_record_count=("subway_record_count", "sum"),
            subway_file_count=("source_file", "nunique")
        )
    )
    out["subway_total_activity"] = out["subway_out_sum"] + out["subway_in_sum"]
    out["subway_abs_net_flow"] = out["subway_net_sum"].abs()
    out["subway_directional_imbalance"] = (
        out["subway_abs_net_flow"] / (out["subway_total_activity"] + 1e-9)
    )

    if period_parts:
        subway_period = pd.concat(period_parts, ignore_index=True)
        subway_period = (
            subway_period.groupby(["MASTER_STATION_ID", "period"], as_index=False)
            .agg(subway_period_activity=("subway_period_activity", "sum"))
        )
        subway_period_wide = subway_period.pivot_table(
            index="MASTER_STATION_ID",
            columns="period",
            values="subway_period_activity",
            aggfunc="sum",
            fill_value=0
        ).reset_index()

        for p in ["morning_peak", "daytime", "evening_peak", "night"]:
            if p not in subway_period_wide.columns:
                subway_period_wide[p] = 0

        subway_period_wide = subway_period_wide.rename(columns={
            "morning_peak": "subway_morning_peak",
            "daytime": "subway_daytime",
            "evening_peak": "subway_evening_peak",
            "night": "subway_night"
        })

        subway_period_wide["subway_period_entropy"] = subway_period_wide[
            ["subway_morning_peak", "subway_daytime", "subway_evening_peak", "subway_night"]
        ].apply(lambda r: entropy_from_values(r.values), axis=1)

    else:
        subway_period_wide = pd.DataFrame()

    return out, subway_period_wide

# ============================================================
# 9. subway 집계
# ============================================================

subway_30, subway_30_period = load_subway_folder(subway_30min_dir, "SUBWAY_30MIN")
subway_daily, subway_daily_period = load_subway_folder(subway_daily_dir, "SUBWAY_DAILY")

subway_30.to_csv(os.path.join(output_dir, "05_subway_30min_all_period.csv"), index=False, encoding="utf-8-sig")
subway_daily.to_csv(os.path.join(output_dir, "06_subway_daily_all_period.csv"), index=False, encoding="utf-8-sig")

subway_all = pd.concat(
    [
        subway_30.assign(source_type="30min"),
        subway_daily.assign(source_type="daily")
    ],
    ignore_index=True
)

subway_total = (
    subway_all.groupby("MASTER_STATION_ID", as_index=False)
    .agg(
        subway_out_sum=("subway_out_sum", "sum"),
        subway_in_sum=("subway_in_sum", "sum"),
        subway_net_sum=("subway_net_sum", "sum"),
        subway_record_count=("subway_record_count", "sum"),
        subway_file_count=("subway_file_count", "sum")
    )
)
subway_total["subway_total_activity"] = subway_total["subway_out_sum"] + subway_total["subway_in_sum"]
subway_total["subway_abs_net_flow"] = subway_total["subway_net_sum"].abs()
subway_total["subway_directional_imbalance"] = (
    subway_total["subway_abs_net_flow"] / (subway_total["subway_total_activity"] + 1e-9)
)

subway_total.to_csv(os.path.join(output_dir, "07_subway_total_all_period.csv"), index=False, encoding="utf-8-sig")

# ============================================================
# 10. 비교 함수: 다양성 feature 포함
# ============================================================

def make_compare(station_mobility, subway_df, label):
    final = station_mobility.merge(subway_df, on="MASTER_STATION_ID", how="inner")

    final["mobility_norm"] = normalize_minmax(final["surrounding_total_flow"])
    final["subway_norm"] = normalize_minmax(final["subway_total_activity"])
    final["mobility_minus_subway_gap"] = final["mobility_norm"] - final["subway_norm"]

    final["hidden_demand_score"] = final["mobility_norm"] * (1 - final["subway_norm"])
    final["subway_overrepresentation_score"] = final["subway_norm"] * (1 - final["mobility_norm"])

    final["distance_norm"] = normalize_minmax(final["mean_distance_m"])
    final["distance_weighted_hidden_demand"] = final["mobility_norm"] * final["distance_norm"]

    final["outer_hidden_demand_score"] = (
        final["hidden_demand_score"] * final["outer_1km_plus_share"]
    )

    final["diverse_living_area_score"] = (
        final["hidden_demand_score"] *
        normalize_minmax(final["distance_band_entropy"]) *
        normalize_minmax(final["period_entropy"])
    )

    final["directional_mismatch_score"] = (
        abs(final["station_directional_imbalance"] - final["subway_directional_imbalance"]) *
        final["mobility_norm"]
    )

    q_hidden = final["hidden_demand_score"].quantile(0.90) if len(final) else 0
    q_over = final["subway_overrepresentation_score"].quantile(0.90) if len(final) else 0
    q_outer = final["outer_hidden_demand_score"].quantile(0.90) if len(final) else 0
    q_diverse = final["diverse_living_area_score"].quantile(0.90) if len(final) else 0
    q_dir = final["directional_mismatch_score"].quantile(0.90) if len(final) else 0

    final["station_type"] = "normal"
    final.loc[final["hidden_demand_score"] >= q_hidden, "station_type"] = "mobility_high_subway_low"
    final.loc[final["subway_overrepresentation_score"] >= q_over, "station_type"] = "subway_high_mobility_low"
    final.loc[final["outer_hidden_demand_score"] >= q_outer, "station_type"] = "outer_hidden_demand"
    final.loc[final["diverse_living_area_score"] >= q_diverse, "station_type"] = "diverse_living_area_hidden"
    final.loc[final["directional_mismatch_score"] >= q_dir, "station_type"] = "directional_mismatch"

    final = final.sort_values("hidden_demand_score", ascending=False)

    out_path = os.path.join(output_dir, f"08_compare_{label}_diverse.csv")
    final.to_csv(out_path, index=False, encoding="utf-8-sig")

    print(f"\n[COMPARE {label}]")
    print(final.shape)
    print(final["station_type"].value_counts())
    print(final[[
        "MASTER_STATION_ID", "nearest_station", "nearest_line",
        "surrounding_total_flow", "subway_total_activity",
        "mobility_minus_subway_gap", "hidden_demand_score",
        "outer_hidden_demand_score",
        "diverse_living_area_score",
        "directional_mismatch_score",
        "station_type"
    ]].head(30))

    return final

# ============================================================
# 11. 비교 실행
# ============================================================

compare_30 = make_compare(station_mobility, subway_30, "subway_30min")
compare_daily = make_compare(station_mobility, subway_daily, "subway_daily")
compare_total = make_compare(station_mobility, subway_total, "subway_total")

# ============================================================
# 12. robust hidden demand
# ============================================================

hidden_30 = set(compare_30.loc[compare_30["hidden_demand_score"] >= compare_30["hidden_demand_score"].quantile(0.90), "MASTER_STATION_ID"])
hidden_daily = set(compare_daily.loc[compare_daily["hidden_demand_score"] >= compare_daily["hidden_demand_score"].quantile(0.90), "MASTER_STATION_ID"])
hidden_total = set(compare_total.loc[compare_total["hidden_demand_score"] >= compare_total["hidden_demand_score"].quantile(0.90), "MASTER_STATION_ID"])

robust_hidden = hidden_30 & hidden_daily & hidden_total

robust_hidden_df = compare_total[
    compare_total["MASTER_STATION_ID"].isin(robust_hidden)
].copy()

robust_hidden_df["robust_hidden_source_count"] = 3

robust_hidden_df.to_csv(
    os.path.join(output_dir, "09_robust_hidden_demand_stations.csv"),
    index=False,
    encoding="utf-8-sig"
)

# ============================================================
# 13. 다양한 관점별 TOP 저장
# ============================================================

top_specs = {
    "10_top_basic_hidden.csv": "hidden_demand_score",
    "11_top_outer_hidden.csv": "outer_hidden_demand_score",
    "12_top_diverse_living_area_hidden.csv": "diverse_living_area_score",
    "13_top_directional_mismatch.csv": "directional_mismatch_score",
    "14_top_subway_overrepresented.csv": "subway_overrepresentation_score",
    "15_top_distance_weighted_hidden.csv": "distance_weighted_hidden_demand",
}

for fname, score_col in top_specs.items():
    out = compare_total.sort_values(score_col, ascending=False).head(50)
    out.to_csv(os.path.join(output_dir, fname), index=False, encoding="utf-8-sig")

# ============================================================
# 14. 요약 리포트 저장
# ============================================================

summary = pd.DataFrame([
    {
        "label": "subway_30min",
        "station_count": len(compare_30),
        "hidden_top10_mean": compare_30["hidden_demand_score"].head(10).mean(),
        "gap_mean": compare_30["mobility_minus_subway_gap"].mean(),
        "gap_median": compare_30["mobility_minus_subway_gap"].median(),
    },
    {
        "label": "subway_daily",
        "station_count": len(compare_daily),
        "hidden_top10_mean": compare_daily["hidden_demand_score"].head(10).mean(),
        "gap_mean": compare_daily["mobility_minus_subway_gap"].mean(),
        "gap_median": compare_daily["mobility_minus_subway_gap"].median(),
    },
    {
        "label": "subway_total",
        "station_count": len(compare_total),
        "hidden_top10_mean": compare_total["hidden_demand_score"].head(10).mean(),
        "gap_mean": compare_total["mobility_minus_subway_gap"].mean(),
        "gap_median": compare_total["mobility_minus_subway_gap"].median(),
    }
])

summary.to_csv(os.path.join(output_dir, "00_analysis_summary.csv"), index=False, encoding="utf-8-sig")

print("\n[ROBUST HIDDEN STATIONS]")
print("count:", len(robust_hidden_df))
print(robust_hidden_df[[
    "MASTER_STATION_ID", "nearest_station", "nearest_line",
    "surrounding_total_flow", "subway_total_activity",
    "hidden_demand_score",
    "outer_hidden_demand_score",
    "diverse_living_area_score",
    "directional_mismatch_score"
]].head(50))

print("\nDONE")
print("saved:", output_dir)
```

```python
import os, glob, re
import pandas as pd
import numpy as np

# ============================================================
# 0. 경로
# ============================================================

mapping_path = r"C:\Users\82108\Desktop\새 폴더 (8)\station_mapping_final\cell_to_nearest_station_VALID_5km_MASTER_ID_FILTERED.csv"

mobility_dir = r"C:\Users\82108\Desktop\새 폴더 (7)\mobility_cell_output"

subway_30min_dir = r"C:\Users\82108\Desktop\새 폴더 (8)\30_od_output (1)\30_od_output"

subway_daily_dir = r"C:\Users\82108\Desktop\새 폴더 (8)\od_output\od_output"

output_dir = r"C:\Users\82108\Desktop\새 폴더 (8)\diverse_mobility_subway_analysis_filtered"
os.makedirs(output_dir, exist_ok=True)

cell_master_path = r"C:\Users\82108\Desktop\새 폴더 (8)\CELL_250x250(UTMK_epsg 5179).csv"
USE_CELL_MASTER_FOR_NUMERIC_BASE = True

# ============================================================
# 1. 유틸
# ============================================================

def read_csv_safe(path):
    for enc in ["utf-8-sig", "cp949", "euc-kr", "utf-8"]:
        try:
            df = pd.read_csv(path, encoding=enc)
            df.columns = df.columns.astype(str).str.strip()
            print("read success:", os.path.basename(path), enc)
            return df
        except UnicodeDecodeError:
            pass
    raise ValueError("CSV 읽기 실패: " + path)

def normalize_minmax(s):
    s = pd.to_numeric(s, errors="coerce").fillna(0)
    if len(s) == 0 or s.max() == s.min():
        return pd.Series(0, index=s.index)
    return (s - s.min()) / (s.max() - s.min())

def is_korean_cell_id(series):
    return series.astype(str).str.contains(r"[가-힣]", regex=True)

def entropy_from_values(values):
    arr = np.array(values, dtype=float)
    arr = arr[arr > 0]
    if len(arr) == 0:
        return 0.0
    p = arr / arr.sum()
    return float(-(p * np.log(p)).sum())

def classify_period(dt):
    h = pd.to_datetime(dt, errors="coerce").dt.hour
    return np.select(
        [
            h.between(6, 9),
            h.between(10, 15),
            h.between(16, 20),
            h.between(21, 23) | h.between(0, 5),
        ],
        ["morning_peak", "daytime", "evening_peak", "night"],
        default="unknown"
    )

# ============================================================
# 2. FILTERED mapping 로드
# ============================================================

mapping = read_csv_safe(mapping_path)

mapping["CELL_ID"] = mapping["CELL_ID"].astype(str).str.strip()
mapping["MASTER_STATION_ID"] = pd.to_numeric(mapping["MASTER_STATION_ID"], errors="coerce")
mapping = mapping.dropna(subset=["MASTER_STATION_ID"]).copy()
mapping["MASTER_STATION_ID"] = mapping["MASTER_STATION_ID"].astype(int)

needed_mapping = [
    "MASTER_STATION_ID", "CELL_ID", "CELL_X", "CELL_Y",
    "nearest_station", "nearest_line",
    "station_x_5179", "station_y_5179",
    "nearest_station_distance_m", "distance_group"
]

missing = [c for c in needed_mapping if c not in mapping.columns]
if missing:
    raise ValueError(f"mapping 필수 컬럼 없음: {missing}")

mapping = (
    mapping
    .sort_values(["CELL_ID", "nearest_station_distance_m", "MASTER_STATION_ID"])
    .drop_duplicates(subset=["CELL_ID"])
    .copy()
)

valid_cell_ids = set(mapping["CELL_ID"].astype(str))
valid_station_ids = set(mapping["MASTER_STATION_ID"].astype(int))

print("\n[MAPPING FILTERED]")
print(mapping.shape)
print("unique cells:", mapping["CELL_ID"].nunique())
print("unique stations:", mapping["MASTER_STATION_ID"].nunique())

# ============================================================
# 3. 숫자형 CELL_ID_BASE 복원용 cell master
# ============================================================

cell_master = None

if USE_CELL_MASTER_FOR_NUMERIC_BASE and os.path.exists(cell_master_path):
    cell_master = read_csv_safe(cell_master_path)
    cell_master.columns = cell_master.columns.astype(str).str.strip()
    cell_master = cell_master.reset_index().rename(columns={"index": "CELL_INDEX"})
    cell_master["CELL_INDEX"] = cell_master["CELL_INDEX"].astype(int)
    cell_master["CELL_ID"] = cell_master["CELL_ID"].astype(str).str.strip()

    # 필요한 cell만 남김
    cell_master = cell_master[cell_master["CELL_ID"].isin(valid_cell_ids)].copy()

    print("\n[CELL MASTER FILTERED]")
    print(cell_master.shape)
else:
    print("\n[CELL MASTER] not used")

# ============================================================
# 4. mobility parquet 읽기 + filtered cell만 사용
# ============================================================

mobility_files = sorted(glob.glob(os.path.join(mobility_dir, "*.parquet")))

print("\n[MOBILITY FILES]", len(mobility_files))

mobility_parts = []
period_parts = []

for i, path in enumerate(mobility_files, 1):
    try:
        df = pd.read_parquet(
            path,
            columns=["datetime", "CELL_ID_BASE", "out_cnt", "in_cnt", "net_cnt"]
        )

        df["CELL_ID_BASE_RAW"] = df["CELL_ID_BASE"].astype(str).str.strip()
        korean_mask = is_korean_cell_id(df["CELL_ID_BASE_RAW"])

        # 한글 CELL_ID
        df_kor = df[korean_mask].copy()
        df_kor["CELL_ID"] = df_kor["CELL_ID_BASE_RAW"]
        df_kor = df_kor[df_kor["CELL_ID"].isin(valid_cell_ids)].copy()

        # 숫자형 CELL_ID_BASE
        if cell_master is not None:
            df_num = df[~korean_mask].copy()
            df_num["CELL_INDEX"] = pd.to_numeric(df_num["CELL_ID_BASE_RAW"], errors="coerce")
            df_num = df_num.dropna(subset=["CELL_INDEX"]).copy()
            df_num["CELL_INDEX"] = df_num["CELL_INDEX"].astype(int)

            df_num = df_num.merge(
                cell_master[["CELL_INDEX", "CELL_ID"]],
                on="CELL_INDEX",
                how="inner"
            )

            df = pd.concat([df_kor, df_num], ignore_index=True)
        else:
            df = df_kor

        if df.empty:
            continue

        for c in ["out_cnt", "in_cnt", "net_cnt"]:
            df[c] = pd.to_numeric(df[c], errors="coerce").fillna(0)

        df["total_flow"] = df["out_cnt"] + df["in_cnt"]
        df["period"] = classify_period(df["datetime"])

        agg = (
            df.groupby("CELL_ID", as_index=False)
            .agg(
                mobility_out_sum=("out_cnt", "sum"),
                mobility_in_sum=("in_cnt", "sum"),
                mobility_net_sum=("net_cnt", "sum"),
                mobility_record_count=("datetime", "count")
            )
        )
        agg["source_file"] = os.path.basename(path)
        mobility_parts.append(agg)

        period_agg = (
            df.groupby(["CELL_ID", "period"], as_index=False)
            .agg(period_total_flow=("total_flow", "sum"))
        )
        period_agg["source_file"] = os.path.basename(path)
        period_parts.append(period_agg)

        if i % 100 == 0:
            print(f"loaded mobility {i}/{len(mobility_files)}")

    except Exception as e:
        print("skip mobility:", os.path.basename(path))
        print("reason:", e)

if not mobility_parts:
    raise ValueError("읽힌 mobility parquet가 없음")

mobility_all = pd.concat(mobility_parts, ignore_index=True)

cell_mobility = (
    mobility_all
    .groupby("CELL_ID", as_index=False)
    .agg(
        mobility_out_sum=("mobility_out_sum", "sum"),
        mobility_in_sum=("mobility_in_sum", "sum"),
        mobility_net_sum=("mobility_net_sum", "sum"),
        mobility_record_count=("mobility_record_count", "sum"),
        mobility_file_count=("source_file", "nunique")
    )
)

cell_mobility["actual_total_flow"] = cell_mobility["mobility_out_sum"] + cell_mobility["mobility_in_sum"]
cell_mobility["actual_abs_net_flow"] = cell_mobility["mobility_net_sum"].abs()
cell_mobility["directional_imbalance"] = (
    cell_mobility["actual_abs_net_flow"] /
    (cell_mobility["actual_total_flow"] + 1e-9)
)

cell_mobility.to_csv(
    os.path.join(output_dir, "01_cell_mobility_filtered.csv"),
    index=False,
    encoding="utf-8-sig"
)

period_all = pd.concat(period_parts, ignore_index=True)

cell_period = (
    period_all.groupby(["CELL_ID", "period"], as_index=False)
    .agg(period_total_flow=("period_total_flow", "sum"))
)

cell_period_wide = (
    cell_period.pivot_table(
        index="CELL_ID",
        columns="period",
        values="period_total_flow",
        aggfunc="sum",
        fill_value=0
    )
    .reset_index()
)

for p in ["morning_peak", "daytime", "evening_peak", "night"]:
    if p not in cell_period_wide.columns:
        cell_period_wide[p] = 0

cell_period_wide["temporal_entropy"] = cell_period_wide[
    ["morning_peak", "daytime", "evening_peak", "night"]
].apply(lambda r: entropy_from_values(r.values), axis=1)

cell_period_wide.to_csv(
    os.path.join(output_dir, "01b_cell_temporal_profile_filtered.csv"),
    index=False,
    encoding="utf-8-sig"
)

print("\n[CELL MOBILITY]")
print(cell_mobility.shape)
print(cell_mobility.head())

# ============================================================
# 5. cell-level merge
# ============================================================

cell_level = mapping.merge(cell_mobility, on="CELL_ID", how="left")
cell_level = cell_level.merge(cell_period_wide, on="CELL_ID", how="left")

fill_cols = [
    "mobility_out_sum", "mobility_in_sum", "mobility_net_sum",
    "mobility_record_count", "mobility_file_count",
    "actual_total_flow", "actual_abs_net_flow",
    "directional_imbalance",
    "morning_peak", "daytime", "evening_peak", "night",
    "temporal_entropy"
]

for c in fill_cols:
    if c in cell_level.columns:
        cell_level[c] = pd.to_numeric(cell_level[c], errors="coerce").fillna(0)

cell_level.to_csv(
    os.path.join(output_dir, "02_cell_level_with_station_filtered.csv"),
    index=False,
    encoding="utf-8-sig"
)

# ============================================================
# 6. station mobility 집계
# ============================================================

station_mobility = (
    cell_level
    .groupby(["MASTER_STATION_ID", "nearest_station", "nearest_line"], as_index=False)
    .agg(
        mapped_cell_count=("CELL_ID", "count"),
        active_cell_count=("actual_total_flow", lambda x: (x > 0).sum()),
        surrounding_total_flow=("actual_total_flow", "sum"),
        surrounding_outflow=("mobility_out_sum", "sum"),
        surrounding_inflow=("mobility_in_sum", "sum"),
        surrounding_netflow=("mobility_net_sum", "sum"),
        mean_cell_flow=("actual_total_flow", "mean"),
        median_cell_flow=("actual_total_flow", "median"),
        max_cell_flow=("actual_total_flow", "max"),
        mean_distance_m=("nearest_station_distance_m", "mean"),
        min_distance_m=("nearest_station_distance_m", "min"),
        max_distance_m=("nearest_station_distance_m", "max"),
        mean_directional_imbalance=("directional_imbalance", "mean"),
        morning_peak_flow=("morning_peak", "sum"),
        daytime_flow=("daytime", "sum"),
        evening_peak_flow=("evening_peak", "sum"),
        night_flow=("night", "sum"),
        mean_temporal_entropy=("temporal_entropy", "mean"),
    )
)

station_mobility["active_cell_ratio"] = (
    station_mobility["active_cell_count"] /
    station_mobility["mapped_cell_count"]
)

station_mobility["station_abs_net_flow"] = station_mobility["surrounding_netflow"].abs()
station_mobility["station_directional_imbalance"] = (
    station_mobility["station_abs_net_flow"] /
    (station_mobility["surrounding_total_flow"] + 1e-9)
)

period_cols = ["morning_peak_flow", "daytime_flow", "evening_peak_flow", "night_flow"]

station_mobility["dominant_period"] = station_mobility[period_cols].idxmax(axis=1)
station_mobility["dominant_period"] = station_mobility["dominant_period"].str.replace("_flow", "", regex=False)

station_mobility["period_entropy"] = station_mobility[period_cols].apply(
    lambda r: entropy_from_values(r.values), axis=1
)

# ============================================================
# 7. 거리권별 집계
# ============================================================

distance_band = (
    cell_level
    .groupby(["MASTER_STATION_ID", "distance_group"], as_index=False)
    .agg(
        band_cell_count=("CELL_ID", "count"),
        band_active_cell_count=("actual_total_flow", lambda x: (x > 0).sum()),
        band_total_flow=("actual_total_flow", "sum"),
        band_mean_flow=("actual_total_flow", "mean"),
        band_max_flow=("actual_total_flow", "max")
    )
)

distance_band.to_csv(
    os.path.join(output_dir, "03_distance_band_station_mobility_filtered.csv"),
    index=False,
    encoding="utf-8-sig"
)

band_wide = distance_band.pivot_table(
    index="MASTER_STATION_ID",
    columns="distance_group",
    values="band_total_flow",
    aggfunc="sum",
    fill_value=0
).reset_index()

for g in ["0-250m", "250-500m", "500m-1km", "1-2km", "2-5km"]:
    if g not in band_wide.columns:
        band_wide[g] = 0

band_cols = ["0-250m", "250-500m", "500m-1km", "1-2km", "2-5km"]

band_wide["distance_band_entropy"] = band_wide[band_cols].apply(
    lambda r: entropy_from_values(r.values), axis=1
)

band_wide["outer_1km_plus_flow"] = band_wide["1-2km"] + band_wide["2-5km"]
band_wide["total_band_flow"] = band_wide[band_cols].sum(axis=1)
band_wide["outer_1km_plus_share"] = (
    band_wide["outer_1km_plus_flow"] /
    (band_wide["total_band_flow"] + 1e-9)
)

station_mobility = station_mobility.merge(
    band_wide[["MASTER_STATION_ID", "distance_band_entropy", "outer_1km_plus_share"]],
    on="MASTER_STATION_ID",
    how="left"
)

station_mobility[["distance_band_entropy", "outer_1km_plus_share"]] = station_mobility[
    ["distance_band_entropy", "outer_1km_plus_share"]
].fillna(0)

station_mobility.to_csv(
    os.path.join(output_dir, "04_station_mobility_diverse_features_filtered.csv"),
    index=False,
    encoding="utf-8-sig"
)

print("\n[STATION MOBILITY]")
print(station_mobility.shape)
print(station_mobility.head())

# ============================================================
# 8. subway 전체 parquet 읽기 함수
# ============================================================

def load_subway_folder(folder, label):
    files = sorted(glob.glob(os.path.join(folder, "*.parquet")))
    print(f"\n[{label} FILES]", len(files))

    parts = []

    for i, path in enumerate(files, 1):
        try:
            df = pd.read_parquet(path)

            if "subwayid" in df.columns and "STATION_ID" not in df.columns:
                df = df.rename(columns={"subwayid": "STATION_ID"})

            needed = ["datetime", "STATION_ID", "od_out_cnt", "od_in_cnt", "od_net_cnt"]
            missing = [c for c in needed if c not in df.columns]

            if missing:
                print("skip:", os.path.basename(path), "missing:", missing)
                continue

            df = df[needed].copy()
            df["STATION_ID"] = pd.to_numeric(df["STATION_ID"], errors="coerce")
            df = df.dropna(subset=["STATION_ID"]).copy()
            df["MASTER_STATION_ID"] = df["STATION_ID"].astype(int)

            # filtered mapping에 있는 station만 사용
            df = df[df["MASTER_STATION_ID"].isin(valid_station_ids)].copy()

            if df.empty:
                continue

            for c in ["od_out_cnt", "od_in_cnt", "od_net_cnt"]:
                df[c] = pd.to_numeric(df[c], errors="coerce").fillna(0)

            df["subway_total_activity_row"] = df["od_out_cnt"] + df["od_in_cnt"]
            df["period"] = classify_period(df["datetime"])

            agg = (
                df.groupby("MASTER_STATION_ID", as_index=False)
                .agg(
                    subway_out_sum=("od_out_cnt", "sum"),
                    subway_in_sum=("od_in_cnt", "sum"),
                    subway_net_sum=("od_net_cnt", "sum"),
                    subway_record_count=("datetime", "count")
                )
            )

            agg["source_file"] = os.path.basename(path)
            parts.append(agg)

            if i % 100 == 0:
                print(f"loaded {label} {i}/{len(files)}")

        except Exception as e:
            print("skip subway:", os.path.basename(path))
            print("reason:", e)

    if not parts:
        return pd.DataFrame()

    all_df = pd.concat(parts, ignore_index=True)

    out = (
        all_df.groupby("MASTER_STATION_ID", as_index=False)
        .agg(
            subway_out_sum=("subway_out_sum", "sum"),
            subway_in_sum=("subway_in_sum", "sum"),
            subway_net_sum=("subway_net_sum", "sum"),
            subway_record_count=("subway_record_count", "sum"),
            subway_file_count=("source_file", "nunique")
        )
    )

    out["subway_total_activity"] = out["subway_out_sum"] + out["subway_in_sum"]
    out["subway_abs_net_flow"] = out["subway_net_sum"].abs()
    out["subway_directional_imbalance"] = (
        out["subway_abs_net_flow"] /
        (out["subway_total_activity"] + 1e-9)
    )

    return out

# ============================================================
# 9. subway 집계
# ============================================================

subway_30 = load_subway_folder(subway_30min_dir, "SUBWAY_30MIN")
subway_daily = load_subway_folder(subway_daily_dir, "SUBWAY_DAILY")

subway_30.to_csv(
    os.path.join(output_dir, "05_subway_30min_filtered.csv"),
    index=False,
    encoding="utf-8-sig"
)

subway_daily.to_csv(
    os.path.join(output_dir, "06_subway_daily_filtered.csv"),
    index=False,
    encoding="utf-8-sig"
)

subway_all = pd.concat(
    [
        subway_30.assign(source_type="30min"),
        subway_daily.assign(source_type="daily")
    ],
    ignore_index=True
)

subway_total = (
    subway_all.groupby("MASTER_STATION_ID", as_index=False)
    .agg(
        subway_out_sum=("subway_out_sum", "sum"),
        subway_in_sum=("subway_in_sum", "sum"),
        subway_net_sum=("subway_net_sum", "sum"),
        subway_record_count=("subway_record_count", "sum"),
        subway_file_count=("subway_file_count", "sum")
    )
)

subway_total["subway_total_activity"] = subway_total["subway_out_sum"] + subway_total["subway_in_sum"]
subway_total["subway_abs_net_flow"] = subway_total["subway_net_sum"].abs()
subway_total["subway_directional_imbalance"] = (
    subway_total["subway_abs_net_flow"] /
    (subway_total["subway_total_activity"] + 1e-9)
)

subway_total.to_csv(
    os.path.join(output_dir, "07_subway_total_filtered.csv"),
    index=False,
    encoding="utf-8-sig"
)

# ============================================================
# 10. 비교 함수
# ============================================================

def make_compare(station_mobility, subway_df, label):
    final = station_mobility.merge(subway_df, on="MASTER_STATION_ID", how="inner")

    final["mobility_norm"] = normalize_minmax(final["surrounding_total_flow"])
    final["subway_norm"] = normalize_minmax(final["subway_total_activity"])
    final["mobility_minus_subway_gap"] = final["mobility_norm"] - final["subway_norm"]

    final["hidden_demand_score"] = final["mobility_norm"] * (1 - final["subway_norm"])
    final["subway_overrepresentation_score"] = final["subway_norm"] * (1 - final["mobility_norm"])

    final["distance_norm"] = normalize_minmax(final["mean_distance_m"])
    final["distance_weighted_hidden_demand"] = final["mobility_norm"] * final["distance_norm"]

    final["outer_hidden_demand_score"] = (
        final["hidden_demand_score"] *
        final["outer_1km_plus_share"]
    )

    final["diverse_living_area_score"] = (
        final["hidden_demand_score"] *
        normalize_minmax(final["distance_band_entropy"]) *
        normalize_minmax(final["period_entropy"])
    )

    final["directional_mismatch_score"] = (
        abs(final["station_directional_imbalance"] - final["subway_directional_imbalance"]) *
        final["mobility_norm"]
    )

    q_hidden = final["hidden_demand_score"].quantile(0.90)
    q_over = final["subway_overrepresentation_score"].quantile(0.90)
    q_outer = final["outer_hidden_demand_score"].quantile(0.90)
    q_diverse = final["diverse_living_area_score"].quantile(0.90)
    q_dir = final["directional_mismatch_score"].quantile(0.90)

    final["station_type"] = "normal"
    final.loc[final["hidden_demand_score"] >= q_hidden, "station_type"] = "mobility_high_subway_low"
    final.loc[final["subway_overrepresentation_score"] >= q_over, "station_type"] = "subway_high_mobility_low"
    final.loc[final["outer_hidden_demand_score"] >= q_outer, "station_type"] = "outer_hidden_demand"
    final.loc[final["diverse_living_area_score"] >= q_diverse, "station_type"] = "diverse_living_area_hidden"
    final.loc[final["directional_mismatch_score"] >= q_dir, "station_type"] = "directional_mismatch"

    final = final.sort_values("hidden_demand_score", ascending=False)

    out_path = os.path.join(output_dir, f"08_compare_{label}_filtered.csv")
    final.to_csv(out_path, index=False, encoding="utf-8-sig")

    print(f"\n[COMPARE {label}]")
    print(final.shape)
    print(final["station_type"].value_counts())

    print(final[[
        "MASTER_STATION_ID", "nearest_station", "nearest_line",
        "surrounding_total_flow", "subway_total_activity",
        "mobility_minus_subway_gap",
        "hidden_demand_score",
        "outer_hidden_demand_score",
        "diverse_living_area_score",
        "directional_mismatch_score",
        "station_type"
    ]].head(30))

    return final

# ============================================================
# 11. 비교 실행
# ============================================================

compare_30 = make_compare(station_mobility, subway_30, "subway_30min")
compare_daily = make_compare(station_mobility, subway_daily, "subway_daily")
compare_total = make_compare(station_mobility, subway_total, "subway_total")

# ============================================================
# 12. robust hidden demand
# ============================================================

hidden_30 = set(compare_30.loc[
    compare_30["hidden_demand_score"] >= compare_30["hidden_demand_score"].quantile(0.90),
    "MASTER_STATION_ID"
])

hidden_daily = set(compare_daily.loc[
    compare_daily["hidden_demand_score"] >= compare_daily["hidden_demand_score"].quantile(0.90),
    "MASTER_STATION_ID"
])

hidden_total = set(compare_total.loc[
    compare_total["hidden_demand_score"] >= compare_total["hidden_demand_score"].quantile(0.90),
    "MASTER_STATION_ID"
])

robust_hidden = hidden_30 & hidden_daily & hidden_total

robust_hidden_df = compare_total[
    compare_total["MASTER_STATION_ID"].isin(robust_hidden)
].copy()

robust_hidden_df["robust_hidden_source_count"] = 3

robust_hidden_df.to_csv(
    os.path.join(output_dir, "09_robust_hidden_demand_stations_filtered.csv"),
    index=False,
    encoding="utf-8-sig"
)

# ============================================================
# 13. 관점별 TOP 저장
# ============================================================

top_specs = {
    "10_top_basic_hidden_filtered.csv": "hidden_demand_score",
    "11_top_outer_hidden_filtered.csv": "outer_hidden_demand_score",
    "12_top_diverse_living_area_hidden_filtered.csv": "diverse_living_area_score",
    "13_top_directional_mismatch_filtered.csv": "directional_mismatch_score",
    "14_top_subway_overrepresented_filtered.csv": "subway_overrepresentation_score",
    "15_top_distance_weighted_hidden_filtered.csv": "distance_weighted_hidden_demand",
}

for fname, score_col in top_specs.items():
    out = compare_total.sort_values(score_col, ascending=False).head(50)
    out.to_csv(
        os.path.join(output_dir, fname),
        index=False,
        encoding="utf-8-sig"
    )

# ============================================================
# 14. 요약 저장
# ============================================================

summary = pd.DataFrame([
    {
        "label": "subway_30min",
        "station_count": len(compare_30),
        "hidden_top10_mean": compare_30["hidden_demand_score"].head(10).mean(),
        "gap_mean": compare_30["mobility_minus_subway_gap"].mean(),
        "gap_median": compare_30["mobility_minus_subway_gap"].median(),
    },
    {
        "label": "subway_daily",
        "station_count": len(compare_daily),
        "hidden_top10_mean": compare_daily["hidden_demand_score"].head(10).mean(),
        "gap_mean": compare_daily["mobility_minus_subway_gap"].mean(),
        "gap_median": compare_daily["mobility_minus_subway_gap"].median(),
    },
    {
        "label": "subway_total",
        "station_count": len(compare_total),
        "hidden_top10_mean": compare_total["hidden_demand_score"].head(10).mean(),
        "gap_mean": compare_total["mobility_minus_subway_gap"].mean(),
        "gap_median": compare_total["mobility_minus_subway_gap"].median(),
    }
])

summary.to_csv(
    os.path.join(output_dir, "00_analysis_summary_filtered.csv"),
    index=False,
    encoding="utf-8-sig"
)

print("\n[ROBUST HIDDEN STATIONS]")
print("count:", len(robust_hidden_df))

print(robust_hidden_df[[
    "MASTER_STATION_ID", "nearest_station", "nearest_line",
    "surrounding_total_flow", "subway_total_activity",
    "hidden_demand_score",
    "outer_hidden_demand_score",
    "diverse_living_area_score",
    "directional_mismatch_score"
]].head(50))

print("\nDONE")
print("saved:", output_dir)
```

```python
import os
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

# ============================================================
# 경로
# ============================================================

input_path = r"C:\Users\82108\Desktop\새 폴더 (8)\diverse_mobility_subway_analysis_filtered\08_compare_subway_total_filtered.csv"

output_dir = r"C:\Users\82108\Desktop\새 폴더 (8)\advanced_station_analysis"
os.makedirs(output_dir, exist_ok=True)

# ============================================================
# 데이터 로드
# ============================================================

df = pd.read_csv(input_path, encoding="utf-8-sig")

print(df.shape)
print(df.columns.tolist())

# ============================================================
# 사용할 feature
# ============================================================

feature_cols = [
    "surrounding_total_flow",
    "subway_total_activity",
    "hidden_demand_score",
    "outer_hidden_demand_score",
    "diverse_living_area_score",
    "directional_mismatch_score",
    "distance_band_entropy",
    "period_entropy",
    "station_directional_imbalance",
    "subway_directional_imbalance",
    "outer_1km_plus_share",
]

# ============================================================
# numeric 변환
# ============================================================

for c in feature_cols:
    df[c] = pd.to_numeric(df[c], errors="coerce")

df = df.dropna(subset=feature_cols).copy()

# ============================================================
# scaling
# ============================================================

X = df[feature_cols].values

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# ============================================================
# KMeans clustering
# ============================================================

N_CLUSTERS = 6

kmeans = KMeans(
    n_clusters=N_CLUSTERS,
    random_state=42,
    n_init=20
)

df["cluster"] = kmeans.fit_predict(X_scaled)

# ============================================================
# cluster summary
# ============================================================

cluster_summary = (
    df.groupby("cluster")[feature_cols]
    .mean()
    .round(4)
)

cluster_summary["station_count"] = (
    df.groupby("cluster").size()
)

cluster_summary = cluster_summary.reset_index()

cluster_summary.to_csv(
    os.path.join(output_dir, "01_cluster_summary.csv"),
    index=False,
    encoding="utf-8-sig"
)

# ============================================================
# cluster별 대표 역
# ============================================================

representatives = []

for c in sorted(df["cluster"].unique()):

    sub = df[df["cluster"] == c].copy()

    sub = sub.sort_values(
        "hidden_demand_score",
        ascending=False
    )

    top = sub.head(20).copy()

    top["cluster"] = c

    representatives.append(top)

representative_df = pd.concat(
    representatives,
    ignore_index=True
)

representative_df.to_csv(
    os.path.join(output_dir, "02_cluster_representative_stations.csv"),
    index=False,
    encoding="utf-8-sig"
)

# ============================================================
# hidden demand corridor 탐색
# ============================================================

corridor = df.copy()

corridor = corridor[
    (corridor["hidden_demand_score"] >= corridor["hidden_demand_score"].quantile(0.85))
    &
    (corridor["outer_1km_plus_share"] >= corridor["outer_1km_plus_share"].quantile(0.70))
].copy()

corridor = corridor.sort_values(
    "hidden_demand_score",
    ascending=False
)

corridor.to_csv(
    os.path.join(output_dir, "03_hidden_demand_corridor.csv"),
    index=False,
    encoding="utf-8-sig"
)

# ============================================================
# directional mismatch 강한 역
# ============================================================

directional = df.sort_values(
    "directional_mismatch_score",
    ascending=False
).copy()

directional.head(100).to_csv(
    os.path.join(output_dir, "04_directional_mismatch_top100.csv"),
    index=False,
    encoding="utf-8-sig"
)

# ============================================================
# 생활권 다양성 강한 역
# ============================================================

diverse = df.sort_values(
    "diverse_living_area_score",
    ascending=False
).copy()

diverse.head(100).to_csv(
    os.path.join(output_dir, "05_diverse_living_area_top100.csv"),
    index=False,
    encoding="utf-8-sig"
)

# ============================================================
# mobility-rich but subway-poor
# ============================================================

rich_poor = df[
    (df["mobility_norm"] >= df["mobility_norm"].quantile(0.80))
    &
    (df["subway_norm"] <= df["subway_norm"].quantile(0.30))
].copy()

rich_poor = rich_poor.sort_values(
    "hidden_demand_score",
    ascending=False
)

rich_poor.to_csv(
    os.path.join(output_dir, "06_mobility_rich_subway_poor.csv"),
    index=False,
    encoding="utf-8-sig"
)

# ============================================================
# subway-rich but mobility-poor
# ============================================================

subway_rich = df[
    (df["subway_norm"] >= df["subway_norm"].quantile(0.80))
    &
    (df["mobility_norm"] <= df["mobility_norm"].quantile(0.30))
].copy()

subway_rich = subway_rich.sort_values(
    "subway_overrepresentation_score",
    ascending=False
)

subway_rich.to_csv(
    os.path.join(output_dir, "07_subway_rich_mobility_poor.csv"),
    index=False,
    encoding="utf-8-sig"
)

# ============================================================
# entropy 기반 unusual station
# ============================================================

unusual = df.copy()

unusual["unusual_score"] = (
    unusual["distance_band_entropy"] *
    unusual["period_entropy"] *
    unusual["directional_mismatch_score"]
)

unusual = unusual.sort_values(
    "unusual_score",
    ascending=False
)

unusual.head(100).to_csv(
    os.path.join(output_dir, "08_unusual_station_patterns.csv"),
    index=False,
    encoding="utf-8-sig"
)

# ============================================================
# 핵심 결과 출력
# ============================================================

print("\n[CLUSTER COUNTS]")
print(df["cluster"].value_counts())

print("\n[TOP HIDDEN DEMAND]")
print(
    df.sort_values(
        "hidden_demand_score",
        ascending=False
    )[
        [
            "nearest_station",
            "nearest_line",
            "hidden_demand_score",
            "outer_hidden_demand_score",
            "diverse_living_area_score",
            "directional_mismatch_score",
            "cluster"
        ]
    ].head(30)
)

print("\n[DIRECTIONAL MISMATCH TOP]")
print(
    directional[
        [
            "nearest_station",
            "nearest_line",
            "directional_mismatch_score",
            "hidden_demand_score"
        ]
    ].head(20)
)

print("\nDONE")
print("saved:", output_dir)
```
